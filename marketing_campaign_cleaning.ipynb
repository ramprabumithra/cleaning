## **IMPORTING LIBRARIES**

```python
import numpy as np
import pandas as pd
```

## **ANALYZING THE DATA**

```python
data = pd.read_csv('marketing_campaign.csv', sep="\t")
print(len(data))
data.head()
```

```python
print(data.shape, data.info())
```

## **CLEANING THE DATASET**

```python
data = data.dropna()
data = data.drop_duplicates()
len(data)
```

```python
data["Dt_Customer"] = pd.to_datetime(data["Dt_Customer"], format='%d-%m-%Y')
```

```python
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_')
data[data.select_dtypes(include='object').columns] = data.select_dtypes(include='object').apply(lambda col: col.str.strip().str.lower())
data.columns
```

```python
data['year_birth'] = data['year_birth'].astype(int)
data['income'] = data['income'].astype(float)
```
